**Task-1**

- I want to create a two lists
- One is Categorcial column list
- Second one is numercial column list
- using dtypes

- Whenever you want create any list from series
- First convert  the series into a Directory
- when you converts it becomes the key:pair value
- Iterate through directory
- if that value === object type, keep that corresponding key in one list
- else keep that variable in other list

```python
data_types=visa_df.dtypes
dict(data_types)
```

```python
cat_list=[]
num_list=[]
# Assigning visa_df.dtypes to datatypes
data_types=visa_df.dtypes
for i,j in dict(data_types).items():
    if j=='object':
        cat_list.append(i)
    else:
        num_list.append(i)
```

```python
cat_list=[ i for i,j in dict(data_types).items() if j=='object']
num_list=[ i for i,j in dict(data_types).items() if j!='object'] 
cat_list,num_list
```

```python
for i,j in dict(data_types).items():
    print(i,j)
```
